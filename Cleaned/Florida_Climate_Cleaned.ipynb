```python
import pandas as pd
import warnings
warnings.simplefilter(action='ignore')
pd.set_option('display.max_rows', 1000)
```

#Creating ds_min

```python
df = pd.read_csv('C:\\Users\\ighos\\Downloads\\Florida Climate\\min_temp.csv', sep='\t')
df = df.astype('string')
df = df.rename(columns = {'                                          POR        JAN   FEB   MAR   APR   MAY   JUN   JUL   AUG   SEP   OCT   NOV   DEC   ANN':'x'})
#dividing dataset into columns
ds = []
for i in range(len(df)):
    ds.append(df.x[i].split(','))
df = pd.DataFrame(ds)
df = df.rename({0:'x', 1:'y'}, axis = 1)
ds = []
for i in range(len(df)):
    ds.append(df.y[i].split())
ds = pd.DataFrame(ds)
#merging all information in one dataset
df = df.drop('y', axis=1)
ds[['x', 'y']] = df['x'].str.extract(r'(\d+)([A-Za-z]+)')
new_order = ['x', 'y'] + [col for col in ds.columns if col not in ['x', 'y']]
ds = ds[new_order]
#choosing only FL cities
ds_min = ds[ds[0] == 'FL']
#rename columns
ds_min = ds_min.rename(columns = {'y':'City', 0:'State', 1:'Observation_Period', 2:'January', 3:'February', 4:'March', 5:'April', 6:'May', 7:'June', 8:'July', 9:'August', 10:'September', 11:'October', 12:'November', 13:'December', 14:'Annual'})
ds_min = ds_min.drop(columns = ['x']).reset_index()
#fixing names of cities
ds_min.City[1] = 'FORT MYERS'
ds_min.City[4] = 'KEY WEST'
ds_min.City[10] = 'VERO BEACH'
ds_min.City[11] = 'WEST PALM BEACH'
#drop useless columns
ds_min = ds_min.drop(columns = ['index', 'State', 'Observation_Period'])
#melt all columns
ds_min = pd.melt(ds_min, id_vars = ['City'], value_vars = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December', 'Annual'] )
ds_min['Type'] = 'Record minimum'
ds_min = ds_min.rename(columns = {'variable':'Month', 'value':'Temperature'})
#enjoy final result
ds_min.to_csv('C:\\Users\\ighos\\Downloads\\Florida Climate\\Cleaned\\ds_min.csv')
ds_min
```

#creating table ds_humidity

```python
#read the csv
df = pd.read_csv('C:\\Users\\ighos\\Downloads\\Florida Climate\\humidity.csv', sep='\t')
df = df.astype('string')
df.columns = ['x']
df = df.drop(index = 0).reset_index()
df = df.drop(columns = ['index'])
ds = []
#split into columns
for i in range(len(df)):
    ds.append(df.x[i].split(','))
df = pd.DataFrame(ds)
df[['x', 'City']] = df[0].str.extract(r'(\d+)([A-Za-z]+)')
df = df.drop(columns = [0,'x'])
df = df[['State', 'x']] = df[1].str.split(n =1, expand = True)
df = df.drop(columns = [1,0])
ds = df.x.str.split(expand = True)
ds.drop(ds.columns[27], axis=1, inplace=True)
ds['State'] = df['State']
#rename columns
ds.columns = ['Years', 'January M', 'January A', 'February M', 'February A', 'March M', 'March A', 'April M', 'April A', 'May M', 'May A', 'June M', 'June A', 'July M', 'July A', 'August M', 'August A', 'September M', 'September A', 'October M', 'October A', 'November M', 'November A', 'December M', 'December A', 'Annual M', 'Annual A', 'State']
ds.drop(columns = ['Years'])
ds = ds[ds.State == 'FL'].reset_index()
ds = ds.drop(columns = ['index'])
#adding cities to ds
ds['City'] = ['DAYTONA BEACH', 'FORT MYERS', 'GAINESVILLE', 'JACKSONVILLE', 'KEY WEST', 'MIAMI', 'ORLANDO', 'PENSACOLA', 'TALLAHASSEE', 'TAMPA', 'VERO BEACH', 'WEST PALM BEACH']
ds_humidity = ds
#melt all columns
ds_humidity = pd.melt(ds_humidity, id_vars = ['City'], value_vars = ['January M','January A','February M','February A','March M','March A','April M','April A','May M','May A','June M','June A','July M','July A','August M','August A','September M','September A','October M','October A','November M','November A','December M','December A','Annual M','Annual A'] )
ds_humidity = ds_humidity.rename(columns = {'variable':'Month', 'value':'Humidity'})
ds_humidity[['Month', 'AM-PM']] = ds_humidity['Month'].str.split(expand = True)
#enjoy final result
ds_humidity.to_csv('C:\\Users\\ighos\\Downloads\\Florida Climate\\Cleaned\\ds_humidity.csv')
ds_humidity
```

#creating table ds_sunshine

```python
#read the csv
df = pd.read_csv('C:\\Users\\ighos\\Downloads\\Florida Climate\\sunshine.csv', sep = '\t')
df = df.astype('string')
#merge the columns into one
df.columns = ['City','y','z','a','b','c','d','e','f','g','h']
df = df.fillna(' ')
df['l'] = df.y+df.z+df.a+df.b+df.c+df.d+df.e+df.f+df.g+df.h
#df.drop(columns = ['y', 'z', 'a', 'b','c','d','e','f','g','h'])
#split columns
df[['Sunshine', 'Days']] = df.l.str.split(n=1, expand = True)
df[['City', 'State']] = df.City.str.split(',', n=1, expand = True)
#drop useless columns
df = df.drop(columns = ['y', 'z', 'a', 'b','c','d','e','f','g','h', 'l'])
df = df.fillna(' ')
ds_sunshine = df[df.State.str.contains('FL')].reset_index()
ds_sunshine = ds_sunshine.drop(columns = ['index'])
ds_sunshine.to_csv('C:\\Users\\ighos\\Downloads\\Florida Climate\\Cleaned\\ds_sunshine.csv')
ds_sunshine
```
